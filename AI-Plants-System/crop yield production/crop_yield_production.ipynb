```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.neighbors import NearestNeighbors
from sklearn.naive_bayes import GaussianNB 
```

```python
ds=pd.read_csv("crop_yield.csv")
```

```python
ds.head()
```

```python
label_encode=LabelEncoder()
ds['Region_new']=label_encode.fit_transform(ds['Region'])
ds['Soil_Type_new']=label_encode.fit_transform(ds['Soil_Type'])
ds['Crop_new']=label_encode.fit_transform(ds['Crop'])
ds['Fertilizer_Used_new']=label_encode.fit_transform(ds['Fertilizer_Used'])
ds['Irrigation_Used_new']=label_encode.fit_transform(ds['Irrigation_Used'])
ds['Weather_Condition_new']=label_encode.fit_transform(ds['Weather_Condition'])
```

```python
x=ds.drop(columns='Yield_tons_per_hectare')
y=ds["Yield_tons_per_hectare"]
x
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
input_data=(2,1,3,992.673282,18.026142,1,1,1,140)
#changing the input data to numpy array
input_data_as_numpy_array=np.array(input_data)
#reshape the array
input_data_reshaped=input_data_as_numpy_array.reshape(1, -1)

prediction=model.predict(input_data_reshaped)
print(prediction)
```
